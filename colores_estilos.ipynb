```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
# plt.style.use('bmh')
# plt.style.use('classic')
# plt.style.use('fivethirtyeight')
plt.style.use('dark_background')
```

## Matlab

```python
fig, axes = plt.subplots(figsize=(8,8))

axes.plot(x,x+1,'r--')
axes.plot(x,x+2,'y-')
axes.plot(x,x+3,'b.-')
axes.plot(x,x+4,'go:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = '#D426C8') # Color RGB
ax.plot(x,x+2,color = '#66FF89')

ax.plot(x,x+3,color = 'blue') # Color comn
ax.plot(x,x+4, color = 'black')
```

```python
fig, ax = plt.subplots(figsize = (6,6)) 
ax.plot(x,x+1,color = '#D426C8', alpha= 0.5, linewidth= 18) #aplha: nivel de transparencia
ax.plot(x,x+2,color = '#66FF89', linewidth= 3) #linewidht = grosor
ax.plot(x,x+3,color = 'blue', linewidth= 5)
ax.plot(x,x+4, color = 'black', alpha = 0.3, linewidth= 12)
```

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = 'x')
ax.plot(x,x+4, color = 'red',linestyle = '--', marker = 'P')
```

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")

ax.plot(x,x+4, color = 'black',linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
```
